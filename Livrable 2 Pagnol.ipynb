## 5. Analyse bivarie

Ici l'on va analyser la relation entre deux variables dans les deux cas numrique-numrique, numrique-catgorielle;

<b>Numrique-Catgorielle</b> : Utilisation de boite  moustaches (box plots) pour visualiser la distribution d'une variable numrique en fonction des catgories d'une variable catgorielle.

<b>Numrique-Numrique</b> : Utilisation d'un heatmap pour visualiser la matrice de corrlation entre les variables numriques.

<b>Catgorielle-Catgorielle</b> :Tables de contingence.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# Identification des variables
# =========================
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
categorical_columns = df.select_dtypes(include=["object"]).columns

print("Colonnes numriques :", numerical_columns)
print("Colonnes catgorielles :", categorical_columns)

# =========================
# 1. Numrique - Catgorielle
# Boxplots
# =========================
# Choisir UNE variable catgorielle de rfrence
cat_ref = categorical_columns[0]

for col in numerical_columns:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=df[cat_ref], y=df[col])
    plt.title(f"{col} selon {cat_ref}")
    plt.xlabel(cat_ref)
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

# =========================
# 2. Numrique - Numrique
# Heatmap de corrlation
# =========================
plt.figure(figsize=(12, 8))
sns.heatmap(
    df[numerical_columns].corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)
plt.title("Matrice de corrlation des variables numriques")
plt.tight_layout()
plt.show()

# =========================
# 3. Catgorielle - Catgorielle
# Crosstab + barres empiles + heatmap
# =========================
if len(categorical_columns) >= 2:
    cat_var_1 = categorical_columns[0]
    cat_var_2 = categorical_columns[1]

    # Tableau de contingence (frquences)
    contingency_table = pd.crosstab(
        df[cat_var_1],
        df[cat_var_2]
    )
    print("\nTableau de contingence :")
    print(contingency_table)

    # Tableau de contingence normalis (proportions)
    contingency_prop = pd.crosstab(
        df[cat_var_1],
        df[cat_var_2],
        normalize="index"
    )

    # Barres empiles
    contingency_prop.plot(
        kind="bar",
        stacked=True,
        figsize=(10, 5)
    )
    plt.title(f"Relation entre {cat_var_1} et {cat_var_2}")
    plt.ylabel("Proportion")
    plt.xlabel(cat_var_1)
    plt.legend(title=cat_var_2)
    plt.tight_layout()
    plt.show()

    # Heatmap des frquences
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        contingency_table,
        annot=True,
        fmt="d",
        cmap="Blues"
    )
    plt.title(f"Heatmap : {cat_var_1} vs {cat_var_2}")
    plt.xlabel(cat_var_2)
    plt.ylabel(cat_var_1)
    plt.tight_layout()
    plt.show()
else:
    print(" Pas assez de variables catgorielles pour une analyse CatCat.")
````

## 3.2 Feature Engineering :

Le feature engineering consiste  crer de nouvelles variables plus explicatives  partir des variables existantes pour amliorer la performance du modle. Elle doivent reflter des aspects pertinents du problme  rsoudre et tre pertinentes pour la tche de prdiction.

Aprs analyse des donnes, nous avons dcid de crer les nouvelles variables suivantes :

### <b>1. TenureRatio :</b>

Les nouvelles variables ont t construites  partir de ratios et de diffrences afin de capturer des dimensions comportementales telles que la stagnation de carrire, la pression de travail perue et la satisfaction globale, facteurs cls de lattrition des employs.

````python
# =========================
# FEATURE ENGINEERING PROPRE
# =========================

# Copie de travail
df_fe = df.copy()

# --- Cration des nouvelles features ---

# 1. Anciennet relative
df_fe['TenureRatio'] = df_fe['YearsAtCompany'] / (df_fe['TotalWorkingYears'] + 1)

# 2. Stagnation de carrire
df_fe['PromotionStagnation'] = (
    df_fe['YearsAtCompany'] - df_fe['YearsSinceLastPromotion']
)

# 3. Pression de travail perue
df_fe['WorkPressure'] = (
    df_fe['AvgWorkTimePerDay_hour'] / (df_fe['WorkLifeBalance'] + 1)
)

# 4. Satisfaction globale
df_fe['GlobalSatisfaction'] = (
    df_fe['EnvironmentSatisfaction'] +
    df_fe['JobSatisfaction'] +
    df_fe['WorkLifeBalance']
) / 3

# 5. Salaire ajust au niveau hirarchique
df_fe['IncomePerLevel'] = (
    df_fe['MonthlyIncome'] / (df_fe['JobLevel'] + 1)
)

# 6. Instabilit professionnelle
df_fe['JobHoppingRate'] = (
    df_fe['NumCompaniesWorked'] / (df_fe['TotalWorkingYears'] + 1)
)

# --- Suppression des colonnes sources ---
df_fe = df_fe.drop(columns=[
    'YearsAtCompany',
    'TotalWorkingYears',
    'YearsSinceLastPromotion',
    'AvgWorkTimePerDay_hour',
    'WorkLifeBalance',
    'EnvironmentSatisfaction',
    'JobSatisfaction',
    'MonthlyIncome',
    'JobLevel',
    'NumCompaniesWorked'
])

# --- Vrification finale ---
print("Colonnes finales :")
print(df_fe.columns)

print("\nStatistiques des nouvelles features :")
print(
    df_fe[
        ['TenureRatio',
         'PromotionStagnation',
         'WorkPressure',
         'GlobalSatisfaction',
         'IncomePerLevel',
         'JobHoppingRate']
    ].describe()
)
````

Un terme correctif de +1 a t ajout aux dnominateurs afin dviter les divisions par zro et de stabiliser les ratios, notamment pour les profils juniors.

### VIF (Variance Inflation Factor) :

Ceci va nous servir  dtecter la multicolinarit entre les variables explicatives. Un VIF lev indique qu'une variable est fortement corrle avec d'autres variables, ce qui peut poser des problmes dans les modles de rgression.

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# =========================
# Slection des variables numriques (sans la cible)
# =========================
X_vif = df_fe.select_dtypes(include=['int64', 'float64']).copy()

# Scurit : retirer Attrition si jamais elle est numrique
if 'Attrition' in X_vif.columns:
    X_vif = X_vif.drop(columns=['Attrition'])

# =========================
# Calcul du VIF
# =========================
vif_data = pd.DataFrame()
vif_data['Feature'] = X_vif.columns
vif_data['VIF'] = [
    variance_inflation_factor(X_vif.values, i)
    for i in range(X_vif.shape[1])
]

# =========================
# Affichage
# =========================
print(vif_data.sort_values(by='VIF', ascending=False))
````

## 4.1 Utilisation de la Rgression Logistique pour Prdire l'Attrition des Employs

### 4.1.1 Qu'est-ce que la Rgression Logistique ? :
La rgression logistique est une technique de modlisation statistique utilise pour prdire la probabilit d'un vnement binaire, tel que le dpart ou la rtention d'un employ dans une entreprise. Contrairement  la rgression linaire qui prdit des valeurs continues, la rgression logistique prdit des probabilits comprises entre 0 et 1.
La rgression logistique utilise la fonction logistique (ou sigmode) pour transformer une combinaison linaire des variables explicatives en une probabilit. La fonction logistique est dfinie comme suit :
$$ P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + ... + \beta_n X_n)}} $$
o :
- \( P(Y=1|X) \) est la probabilit que l'vnement d'intrt (par exemple, le dpart de l'employ) se produise, tant donn les variables explicatives \( X \).
- \( \beta_0 \) est l'ordonne  l'origine (intercept).

## 4.2 Sparation des donnes en 80% Train (Apprentissage) et 20% (pour le test)

Cette tape consiste a divis notre dataset en deux partie 80%  donn  notre modle pour l'apprentissage et ensuite l'on va test avec les 20% restant.

Pour evit les biais donc les erreurs le dataset est divis de maniere latoire grace au StarifiedShuffleSplit

````python
from sklearn.model_selection import StratifiedShuffleSplit

# =========================
# Sparation features / cible sur le dataset final
# =========================
X = df_fe.drop(columns=['Attrition'])
y = df_fe['Attrition']

# =========================
# Stratified Shuffle Split
# =========================
sss = StratifiedShuffleSplit(
    n_splits=1,        # un seul split
    test_size=0.2,     # 20% test
    random_state=42    # reproductibilit
)

# Extraction des indices
for train_idx, test_idx in sss.split(X, y):
    X_train = X.iloc[train_idx].copy()
    X_test  = X.iloc[test_idx].copy()
    y_train = y.iloc[train_idx].copy()
    y_test  = y.iloc[test_idx].copy()

# =========================
# Vrifications
# =========================
print("Dimensions :")
print(f"X_train : {X_train.shape}")
print(f"X_test  : {X_test.shape}")
print(f"y_train : {y_train.shape}")
print(f"y_test  : {y_test.shape}")

print("\nProportion de la variable cible (Attrition) :")
print("Train :")
print(y_train.value_counts(normalize=True))
print("\nTest :")
print(y_test.value_counts(normalize=True))
````

### 4.3 Sparation des variables explicatives et de la variable cible

````python
from sklearn.model_selection import StratifiedShuffleSplit

X = df_final.drop(columns=['Attrition'])
y = df_final['Attrition']

sss = StratifiedShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=42
)

for train_idx, test_idx in sss.split(X, y):
    X_train = X.iloc[train_idx]
    X_test  = X.iloc[test_idx]
    y_train = y.iloc[train_idx]
    y_test  = y.iloc[test_idx]

print("Train shape:", X_train.shape)
print("Test shape :", X_test.shape)
print("\nAttrition ratio (train):")
print(y_train.value_counts(normalize=True))
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# =========================
# Modle de base
# =========================
log_reg = LogisticRegression(
    max_iter=2000,
    random_state=42
)

# =========================
# Grille d'hyperparamtres
# =========================
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear'],
    'class_weight': [None, 'balanced']
}

# =========================
# GridSearch
# =========================
grid = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1,
    verbose=1
)

# =========================
# Entranement
# =========================
grid.fit(X_train, y_train)

# =========================
# Meilleur modle
# =========================
best_log_reg = grid.best_estimator_

print("Meilleurs hyperparamtres :")
print(grid.best_params_)

print("\nMeilleur ROC-AUC (CV) :")
print(grid.best_score_)
````

## INDICATEURS DE PERFORMANCE DU MODELE : 

Une fois le modle entrain, nous allons valuer ses performances en utilisant plusieurs mtriques cls adaptes aux problmes de classification binaire, notamment :
- Exactitude (Accuracy) : Mesure la proportion de prdictions correctes parmi l'ensemble des prdictions effectues. C'est une mtrique globale qui donne une ide gnrale de la performance du modle.
- Prcision (Precision) : Indique la proportion de prdictions positives correctes parmi toutes les prdictions positives effectues par le modle. Elle est particulirement importante lorsque le cot des fausses alertes est lev.
- Rappel (Recall) : Mesure la proportion de vrais positifs correctement identifis parmi tous les cas positifs rels. C'est crucial lorsque le cot des faux ngatifs est lev.
- F1-Score : La moyenne harmonique entre la prcision et le rappel, offrant un quilibre entre les deux mtriques.
- AUC-ROC (Area Under the Receiver Operating Characteristic Curve) : Mesure la capacit du modle  distinguer entre les classes positives et ngatives  diffrents seuils de classification. courbe ROC, qui trace le taux de vrais positifs contre le taux de faux positifs  diffrents seuils.


````python
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# Prdictions
# =========================
y_pred = best_log_reg.predict(X_test)
y_proba = best_log_reg.predict_proba(X_test)[:, 1]

# =========================
# Matrice de confusion
# =========================
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# =========================
# Rapport prcision / rappel / F1
# =========================
print("Classification Report :")
print(classification_report(y_test, y_pred))

# =========================
# ROC-AUC
# =========================
auc = roc_auc_score(y_test, y_proba)
print("ROC-AUC :", auc)

# =========================
# Courbe ROC
# =========================
fpr, tpr, _ = roc_curve(y_test, y_proba)

plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
````
