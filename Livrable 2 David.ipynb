```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

### 2.1.2. Chargement des fichiers

Les diffrents fichiers fournis sont rassembls et fusionns  laide de lidentifiant unique EmployeeID, permettant de relier les informations RH, valuations managriales, enqutes de satisfaction et horaires de travail. Cette consolidation garantit un jeu de donnes complet et cohrent, indispensable pour identifier prcisment les facteurs influenant lattrition.

```python
# Charger le dataset
employee_survey_data = pd.read_csv('datasets\\employee_survey_data.csv')
general_data = pd.read_csv('datasets\\general_data.csv')
manager_survey_data = pd.read_csv('datasets\\manager_survey_data.csv')
in_time = pd.read_csv('datasets\\in_time.csv')
out_time = pd.read_csv('datasets\\out_time.csv')

df = pd.merge(employee_survey_data, general_data, on = 'EmployeeID', how = 'inner')
df = pd.merge(df, manager_survey_data, on = 'EmployeeID', how = 'inner')

# Suppression de la colonne EmployeeID
df =df.drop(columns=['EmployeeID'], inplace=False)

# Conversion des dates et heures en type datetime
in_time = in_time.apply(pd.to_datetime, errors='coerce')
out_time = out_time.apply(pd.to_datetime, errors='coerce')

# Calcul du temps de travail
work_time = out_time - in_time

# Ajouter les variables au dataframe principal
df['AvgWorkTimePerDay_hour'] = work_time.mean(axis=1, skipna=True).dt.total_seconds() / 3600


# Affichage des premires lignes du dataset fusionn
df.head()


```

## 2. Statistique descriptive

`La statistique descriptive` consiste  **rsumer et explorer les donnes** afin de mieux comprendre leur structure avant toute modlisation. Pour ce projet, elle permet de dtecter les tendances gnrales, les anomalies, les valeurs manquantes et de mieux apprhender le comportement des employs vis--vis de lattrition.  

Elle inclut :  
- **Mesures de tendance centrale** : moyenne, mdiane, minimum et maximum des variables numriques (ge, salaire, temps de travail).  
- **Mesures de dispersion** : cart-type, quartiles et distribution des donnes pour identifier la variabilit entre employs.  
- **Variables catgorielles** : nombre demploys par poste, par niveau hirarchique, par statut marital, etc.  
- **Corrlations** : relations entre variables numriques pour comprendre quelles caractristiques pourraient influencer le dpart des employs.  
- **Visualisation des donnes** : histogrammes, boxplots, barplots et heatmaps pour rendre les informations facilement interprtables.  


```python
# Affichage complet des colonnes
pd.set_option('display.max_columns', None)

# ---- Conversion Yes/No en numrique pour certaines colonnes ----
yes_no_cols = ['Over18']  # Ajouter d'autres si ncessaire
for col in yes_no_cols:
    if col in df.columns:
        df[col] = df[col].map({'Yes': 1, 'No': 0})

# ---- Colonnes numriques relles ----
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# ---- Statistique descriptive ----
numeric_summary = df[numeric_cols].describe()
print("### Statistiques descriptives (numriques)")
display(numeric_summary)

# ---- Statistiques descriptives pour les variables catgorielles ----
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
categorical_summary = df[categorical_cols].describe()
print("### Statistiques descriptives (catgorielles)")
display(categorical_summary)

# ---- Valeurs manquantes ----
missing_values = df.isnull().sum()
print("### Valeurs manquantes par colonne")
display(missing_values)

# ---- Corrlation entre variables numriques ----
correlation = df[numeric_cols].corr()
print("### Corrlations entre variables numriques")
display(correlation)

# ---- Visualisations simples ----
# Histogrammes
df[numeric_cols].hist(figsize=(15,10), bins=20)
plt.suptitle("Histogrammes des variables numriques")
plt.show()

# Boxplot exemple
plt.figure(figsize=(10,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Salaire brut par attrition")
plt.show()

# Barplot exemple
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole', data=df, order=df['JobRole'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Rpartition des employs par poste")
plt.show()

# Heatmap corrlations
plt.figure(figsize=(12,10))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Corrlation entre variables numriques")
plt.show()
```
